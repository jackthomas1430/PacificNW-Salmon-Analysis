```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the temperature data
temperature_file_path = 'Nicoles_temperature_data.xlsx'
temperature_data = pd.read_excel(temperature_file_path, sheet_name='New_temperature_data')

# Filter data for the "Interior Columbia" region
interior_columbia_data = temperature_data[temperature_data['REGION'] == 'Interior Columbia']

# Filter for days where the DAILY_MEAN_WATER_TEMP is less than 14°C
below_14_degrees = interior_columbia_data[interior_columbia_data['DAILY_MEAN_WATER_TEMP'] < 14]

# Convert the 'DATE' column to datetime format
below_14_degrees['DATE'] = pd.to_datetime(below_14_degrees['DATE'])

# Extract the year from the 'DATE' column
below_14_degrees['YEAR'] = below_14_degrees['DATE'].dt.year

# Filter the data to include only years up to 2021
below_14_degrees = below_14_degrees[below_14_degrees['YEAR'] <= 2021]

# Group by SITE NAME and YEAR, counting the occurrences
yearly_counts = below_14_degrees.groupby(['SITE NAME', 'YEAR']).size().reset_index(name='Number of Days Below 14°C')

# Pivot the data to have years as index and site names as columns
pivot_data = yearly_counts.pivot(index='YEAR', columns='SITE NAME', values='Number of Days Below 14°C').fillna(0)

# Plot the line graph for each site
plt.figure(figsize=(10, 6))

for site in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[site], marker='o', label=site)
    
    # Annotate each point with the number of days
    for year, value in pivot_data[site].items():
        plt.text(year, value, f'{int(value)}', ha='center', va='bottom')

plt.xlabel('Year')
plt.ylabel('Number of Days Below 14°C')
plt.title('Yearly Number of Days with DAILY_MEAN_WATER_TEMP Below 14°C for Interior Columbia Region (Up to 2021)')
plt.legend(title='Site Name')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the population data
population_file_path = 'Nicoles_population_data.xlsx'
population_data = pd.read_excel(population_file_path, sheet_name='New Population Data')

# Filter data for the "Clearwater River Lower Mainstem" population
clearwater_data = population_data[population_data['POPULATION_NAME'] == 'Clearwater River Lower Mainstem']

# Group by year and sum the NUMBER_OF_SPAWNERS
yearly_spawners = clearwater_data.groupby('YEAR')['NUMBER_OF_SPAWNERS'].sum().reset_index()

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(yearly_spawners['YEAR'], yearly_spawners['NUMBER_OF_SPAWNERS'], marker='o', color='b')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sum of Number of Spawners')
plt.title('Yearly Sum of Number of Spawners for Clearwater River Lower Mainstem')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the temperature data
temperature_file_path = 'Nicoles_temperature_data.xlsx'
temperature_data = pd.read_excel(temperature_file_path, sheet_name='New_temperature_data')

# Filter for days where the DAILY_MEAN_WATER_TEMP is less than 14°C
below_14_degrees = temperature_data[temperature_data['DAILY_MEAN_WATER_TEMP'] < 14]

# Convert the 'DATE' column to datetime format
below_14_degrees['DATE'] = pd.to_datetime(below_14_degrees['DATE'])

# Extract the year from the 'DATE' column
below_14_degrees['YEAR'] = below_14_degrees['DATE'].dt.year

# Group by RUN_TIME and YEAR, counting the occurrences of days below 14°C
yearly_counts_by_runtime = below_14_degrees.groupby(['Run-Time', 'YEAR']).size().reset_index(name='Number of Days Below 14°C')

# Pivot the data to have years as index and RUN_TIME as columns
pivot_data = yearly_counts_by_runtime.pivot(index='YEAR', columns='Run-Time', values='Number of Days Below 14°C').fillna(0)

# Plot the line graph for each RUN_TIME (season)
plt.figure(figsize=(10, 6))

for season in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[season], marker='o', label=season)
    
    # Annotate each point with the number of days
    for year, value in pivot_data[season].items():
        plt.text(year, value, f'{int(value)}', ha='center', va='bottom')

plt.xlabel('Year')
plt.ylabel('Number of Days Below 14°C')
plt.title('Yearly Number of Days with DAILY_MEAN_WATER_TEMP Below 14°C Broken Down by Run-Time (Seasons)')
plt.legend(title='Run-Time (Season)')
plt.tight_layout()
plt.show()
```
