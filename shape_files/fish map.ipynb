```python
#update coordinates

# Load the existing points shapefile
points_gdf = gpd.read_file('points/points_data.shp')

# Define the target region and species to update
target_region = 'Willamette/Lower Columbia'
species_to_update = ['STEELHEAD', 'CHINOOK SALMON', 'CHUM SALMON']

# Filter the points for the target region and species
filtered_points = points_gdf[(points_gdf['Region'] == target_region) &
                              (points_gdf['Species'].isin(species_to_update))]

# Define new coordinates for specific points
new_coordinates = {
    ('STEELHEAD', 22): (45.53638, -122.58111),
    ('STEELHEAD', 23): (45.53638, -122.58111),
    ('STEELHEAD', 24): (45.53638, -122.58111),
    ('STEELHEAD', 25): (45.53638, -122.58111),
    ('STEELHEAD', 26): (45.53638, -122.58111),
    ('STEELHEAD', 27): (45.53638, -122.58111),
    ('STEELHEAD', 28): (45.53638, -122.58111),
    ('CHINOOK SALMON', 15): (45.08972, -122.87249),
    ('CHINOOK SALMON', 16): (45.08972, -122.87249),
    ('CHINOOK SALMON', 17): (45.08972, -122.87249),
    ('CHINOOK SALMON', 18): (45.08972, -122.87249),
    ('CHINOOK SALMON', 19): (45.08972, -122.87249),
    ('CHINOOK SALMON', 20): (45.08972, -122.87249),
    ('CHINOOK SALMON', 21): (45.08972, -122.87249),
    ('CHUM SALMON', 29): (44.40222, -123.04055),
    ('CHUM SALMON', 30): (44.40222, -123.04055),
    ('CHUM SALMON', 31): (44.40222, -123.04055),
    ('CHUM SALMON', 32): (44.40222, -123.04055),
    ('CHUM SALMON', 33): (44.40222, -123.04055),
    ('CHUM SALMON', 34): (44.40222, -123.04055)
}

# Apply the updates to the GeoDataFrame
def update_coordinates(row):
    key = (row['Species'], row['ID'])  # Adjust 'ID' to your unique identifier column
    if key in new_coordinates:
        new_lat, new_lon = new_coordinates[key]
        row.geometry = gpd.points_from_xy([new_lon], [new_lat])[0]
    return row

points_gdf_updated = points_gdf.apply(update_coordinates, axis=1)

# Overwrite the existing GeoDataFrame file with updated data
points_gdf_updated.to_file('points/points_data.shp')
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objects as go
import geopandas as gpd
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash

# Load and prepare data
points_gdf_updated = gpd.read_file('points/points_data.shp')
points_gdf_updated = points_gdf_updated.to_crs(epsg=4326)  # Ensure using a common CRS (WGS84)

polygons_gdf = gpd.read_file('polygons/polygon_data.shp')
polygons_gdf = polygons_gdf.to_crs(epsg=4326)  # Ensure using a common CRS (WGS84)

# Initialize the Dash app
app = JupyterDash(__name__)

# Layout of the Dash app
app.layout = html.Div([
    html.H1("Oregon Salmon Populations 2013-2019"),
    
    # Dropdown for selecting year
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': year, 'value': year} for year in range(2013, 2020)],
        value=2013  # Default value
    ),
    
    # Graph for map
    dcc.Graph(id='map-plot')
])

# Callback to update map based on dropdown selection
@app.callback(
    Output('map-plot', 'figure'),
    [Input('year-dropdown', 'value')]
)
def update_map(selected_year):
    # Filter data based on selected year
    filtered_points_df = points_gdf_updated[points_gdf_updated['Year'] == selected_year]
    
    # Define a color map for species
    color_map = {
        'CHINOOK SALMON': 'blue',
        'STEELHEAD': 'green',
        'CHUM SALMON': 'orange',
        'COHO SALMON': 'purple'
    }
    
    # Define a color map for region boundaries
    region_color_map = {
        'Willamette/Lower Columbia': 'yellow',
        'Oregon Coast': 'lightgreen',
        'Interior Columbia': 'lightblue'
    }
    
    # Create a plotly figure
    fig = go.Figure()

    # Add colored polygons with boundaries
    for _, row in polygons_gdf.iterrows():
        region = row['Region']
        fig.add_trace(go.Scattermapbox(
            lon=[p[0] for p in row.geometry.exterior.coords],
            lat=[p[1] for p in row.geometry.exterior.coords],
            mode='lines',
            line=dict(width=2, color=region_color_map.get(region, 'grey')),  # Boundary color
            fillcolor='rgba(0,0,0,0)',  # No fill color
            fill='none',
            name=region,  # Only add one entry per region
            text=region,  # Text to display on hover
            hoverinfo="text",
            opacity=1,  # Fully opaque boundaries
            showlegend=False  # Hide actual region polygons from the legend
        ))

    # Add labels for polygons
    for _, row in polygons_gdf.iterrows():
        fig.add_trace(go.Scattermapbox(
            lat=[row.geometry.centroid.y],
            lon=[row.geometry.centroid.x],
            text=[row['Region']],
            mode='text',
            textposition='middle center',
            showlegend=False,
            textfont=dict(size=12, color='black', family='Arial')  # Make text bold
        ))

    # Add points
    for species, color in color_map.items():
        species_points = filtered_points_df[filtered_points_df['Species'] == species]
        fig.add_trace(go.Scattermapbox(
            lat=species_points.geometry.y,
            lon=species_points.geometry.x,
            mode='markers+text',
            text=species_points['Species'],
            textposition='top right',
            marker=dict(size=12, color=color, opacity=0.8),
            hovertemplate='<b>%{text}</b><br>Population: %{customdata[0]}<extra></extra>',
            customdata=species_points[['Population']].values,
            showlegend=False  # Hide legend for salmon species
        ))

    # Add dummy traces for the legend to show color labels
    for region_name, color in region_color_map.items():
        fig.add_trace(go.Scattermapbox(
            lat=[None],  # No actual data
            lon=[None],
            mode='markers',
            marker=dict(size=10, color=color),
            name=f'{color} = {region_name}',
            showlegend=True  # Show these entries in the legend
        ))

    # Set map style and layout
    fig.update_layout(
        mapbox_style="carto-positron",
        mapbox=dict(center=dict(lat=45.0, lon=-120.0), zoom=5),
        title=f'Fish Species for the Year {selected_year}',
        showlegend=True,  # Show legend
        legend=dict(
            title="Regions",
            title_font=dict(size=14),
            font=dict(size=12),
            traceorder='normal',
            itemsizing='constant',
            itemclick='toggleothers'
        )
    )
    
    return fig

# Run the app
app.run_server(mode='external', debug=True)
```
